```python
# -*- coding: utf-8 -*-

from matplotlib import pyplot as plt

h = [1, 1+(0.9), 1+(0.9)+(0.9)**2]
for i in range(3, 51) :
    h.append(0.9 * h[i-1])

plt.xlabel('n')
plt.ylabel('h[n]')
plt.bar(range(len(h)), h, width=0.1, color="red")
plt.title("201710860 h[n] drawing")
plt.show()
```

```python
from matplotlib import pyplot as plt

# x[n] = delta[n]
X = [0 for i in range(51)]
X[0] = 1

Y = [0 for i in range(51)]
Y[0] = 1
for i in range(1, 51) :
    # X[-1] == X[50] == 0   
    Y[i] = 0.9*Y[i-1] + X[i] + X[i-1] + X[i-2]

plt.xlabel('n')
plt.ylabel('Y[n]')
plt.bar(range(51), Y, width=0.1, color="red")
plt.title("201710860 Y[n] drawing")
plt.show()
```
